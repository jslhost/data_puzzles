# Intro

**Every minute**, the spider moves at an angle taken at random **uniformly between 0 and 360degrees**, at a distance taken at random **uniformly between 0 and 2m**. 

The spider starts at coordinates (0, 0) and my place can be represented by the area determined by the square delimited by upper-left corner (10, 10) and lower-right corner (15, 6) (unit is meter)

What's the probability the spider is at my place after 240mn?

# Solve

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

def spider_simulation(nb_simulations=100):
    trajectory = []

    for _ in range(nb_simulations):
        x = y = 0
        N_MINUTES = 240

        for i in range(N_MINUTES):
            angle = np.random.uniform(0, 365, 1)
            distance = np.random.uniform(0, 2, 1)
            x = x + np.cos(angle) * distance
            y = y + np.sin(angle) * distance

        spider_coordinates = (x[0], y[0])
        trajectory.append(spider_coordinates)
    return trajectory
```

```python
trajectory = spider_simulation()
plt.plot((10, 10, 15, 15, 10), (10, 6, 6, 10, 10), c='r')

for coordinate in trajectory:
    plt.plot((0, coordinate[0]), (0, coordinate[1]))
    plt.scatter((0, coordinate[0]), (0, coordinate[1]))
```

```python
counter = 0
nb_simulations = 10_000
trajectories = spider_simulation(nb_simulations)

for trajectory in trajectories:
    if 10 <= trajectory[0] <= 15 and 6 <= trajectory[1] <= 10:
        counter += 1

proba = counter / nb_simulations
print(round(proba * 100, 1), '%')
```
