# Problem summary


Can we extract the birth date of that person from that set of data?

*Some starter code:*



```python
import pandas as pd
import matplotlib.pyplot as plt

#Â Load data
dd_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/dd.csv')
mm_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/mm.csv')
yyyy_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/yyyy.csv')
```

```python
# Plot the DD data
plt.scatter(dd_df.x, dd_df.y)
```

```python
# Plot the MM data
plt.scatter(mm_df.x, mm_df.y, alpha=0.1)
```

```python
# Plot the YYYY data
plt.scatter(yyyy_df.x, yyyy_df.y)
```

# Your turn

Find the birth date 'encrypted' in these datasets!

```python
import numpy as np
from sklearn.linear_model import LinearRegression

model = LinearRegression()
dd_x = dd_df['x'].values.reshape(-1, 1)
dd_y = dd_df['y'].values.reshape(-1, 1)

model.fit(dd_x, dd_y)

print(model.coef_, model.intercept_)

f = lambda x :(model.coef_[0] * x) + model.intercept_[0]
x = np.linspace(-150, 150)
plt.plot(x, f(x), c='r')
plt.scatter(dd_df.x, dd_df.y)
plt.show()
```

```python
model = LinearRegression()
mm_x = mm_df['x'].values.reshape(-1, 1)
mm_y = mm_df['y'].values.reshape(-1, 1)

model.fit(mm_x, mm_y)

print(model.coef_, model.intercept_)

f = lambda x :(model.coef_[0] * x) + model.intercept_[0]
x = np.linspace(-150, 150)
plt.plot(x, f(x), c='r')
plt.scatter(mm_df.x, mm_df.y, alpha=0.1)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
model = LinearRegression()
poly = PolynomialFeatures(3)
yyyy_x = poly.fit_transform(yyyy_df['x'].values.reshape(-1, 1))
yyyy_y = yyyy_df['y'].values.reshape(-1, 1)

model.fit(yyyy_x, yyyy_y)

print(model.coef_[0][::-1], model.intercept_)

f = lambda x :(model.coef_[0][3] * x**3) + (model.coef_[0][2] * x**2) + (model.coef_[0][1] * x) + model.intercept_[0]
x = np.linspace(-550, 550)
plt.plot(x, f(x), c='r')
plt.scatter(yyyy_df.x, yyyy_df.y)
plt.show()
```
